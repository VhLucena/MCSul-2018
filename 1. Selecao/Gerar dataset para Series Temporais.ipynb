## Objetivo:
* Gerar um dataframe com a quantidade de acidentes por dia entre 2000 e 2016

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ano = 2000
dados = pd.DataFrame()
while (ano < 2017):
    if(ano!=2015):
        filename = "../Database/acidentes_" + str(ano) + ".csv"
        df = pd.read_csv(filename, sep=";")
        dados = pd.concat([dados, df[['DIA','MES','ANO']]])    
    ano += 1

```

```python
serie = []
ano = 2000
while (ano <= 2016):
    for mes in range(13):
        for dia in range(32):
            cont = dados[(dados['DIA']==dia) & (dados['MES']==mes) & (dados['ANO']==ano)]['DIA'].count()
            if (cont != 0):
                serie.append(cont)
    ano += 1
    
```
