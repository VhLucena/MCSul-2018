# 2. Pre-processamento

## Objetivo
* Incluir o rótulo nos dados
* Discretizar atributo 'hora' em Manha, Tarde, Noite e Madrugada

## Técnicas aplicadas
*  2.1. Combinar dois atributos
* 2.2. Discretizar atributo 'hora'
* 2.2.1. Dados faltantes


------

```python
import pandas as pd
import math
import matplotlib.pyplot as plt
```

### 2.1 Combinar dois atributos 
* FATAIS e FERIDOS_GR => ACID_GRAVE

```python
dfOriginal = pd.read_csv("../Database/1.Dados_Selecionados/Round2_dados_selecionados.csv")
```

```python
dfAcidentesGraves = dfOriginal[(dfOriginal['FATAIS'] > 0) | (dfOriginal['FERIDOS_GR'] > 0)]
dfAcidentesGraves['ACID_GRAVE'] = ['SIM' for i in range(len(dfAcidentesGraves))]
```

```python
dfAcidentesNormais = dfOriginal[(dfOriginal['FATAIS'] == 0) & (dfOriginal['FERIDOS_GR'] == 0)]
dfAcidentesNormais['ACID_GRAVE'] = 'NAO' 
```

```python
dfTotal = pd.concat([dfAcidentesGraves, dfAcidentesNormais])
dfTotal.drop(columns=['FATAIS', 'FERIDOS_GR'], inplace=True)
```

```python
dfTotal.dropna(inplace=True)
```

### 2.2 Discretizar Atributo 'hora'

## Considerações:
* 06h..<= **manhã**......< 12h
* 12h..<= **tarde**......< 18h
* 18h..<= **noite**......< 00h
* 00h..<= **madrugada**..< 6h

```python
horas = dfTotal['HORA']
```

----

#### 2.2.1. Dados faltantes
* Solução: Remover o registro com dados faltantes

----

```python
horas[(horas>='06:00') & (horas<='12:00')] = 'MANHA'
horas[(horas>='12:00') & (horas<='18:00')] = 'TARDE'
horas[(horas>='18:00') & (horas<='24:00')] = 'NOITE'
horas[(horas>='00:00') & (horas<='06:00')] = 'MADRUGADA'
```

```python
dfTotal.drop(columns='HORA_DISCRETIZADA', inplace=True)
```

### 2.2 Excluir Atributo 'hora'

```python
# Excluindo atributo hora
dfTotal.drop(columns='HORA', axis=1, inplace=True)
```
